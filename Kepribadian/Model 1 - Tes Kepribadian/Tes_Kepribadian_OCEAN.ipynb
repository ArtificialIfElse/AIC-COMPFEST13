## 5. Membuat Model

### 5.1 Membuat model

```python
# Mmembuat cluster K-Means model
from sklearn.cluster import KMeans

# Membackup data df
df_model = df.copy()

# Membuat 5 kluster dan melakukan fitting pada model
kmeans = KMeans(n_clusters=5)
k_fit = kmeans.fit(df_model)
```

```python
# Melakukan prediksi kluster pada kolom baru
pd.options.display.max_columns = 10 # Mengurangi maksimal kolom yang ditampilkan menjadi 10
predictions = k_fit.labels_
df_model['Clusters'] = predictions
df_model.head()
```

### 5.2 Menganalisis kluster model

```python
# Menghitung jumlah masing-masing kluster
df_model.Clusters.value_counts()
```

```python
pd.options.display.max_columns = 150
df_model.groupby('Clusters').mean()
```

```python
df_model.groupby('Clusters').mean().round() + 1
```

* Extroversion : (I)ntroverted >< (E)xtroverted
* Openness : ob(S)ervant >< i(N)tuitive
* Conscientiousness : (J)udging >< (P)erceiving
* Agreeableness : (T)hinking >< (F)eeling
* Neuroticism : (A)ssertive >< (T) Turbulant

```python
df_model.head()
```

```python
# Menjumlah setiap pertanyaan untuk melihat pattern dari masing-masing cluster
col_list = list(df_model)
ext = col_list[0:10]
est = col_list[10:20]
agr = col_list[20:30]
csn = col_list[30:40]
opn = col_list[40:50]

df_total = pd.DataFrame()
df_total['extroversion'] = df_model[ext].sum(axis=1)/10
df_total['openness'] = df_model[opn].sum(axis=1)/10
df_total['conscientiousness'] = df_model[csn].sum(axis=1)/10
df_total['agreeableness'] = df_model[agr].sum(axis=1)/10
df_total['neuroticism'] = df_model[est].sum(axis=1)/10

df_total['clusters'] = predictions
df_total.groupby('clusters').mean()
```

```python
# Memvisualisasikan rata-rata pertanyaan tiap grup
df_clusters = df_total.groupby('clusters').mean()
plt.figure(figsize = (22, 3))
for i in range(0,5):
    plt.subplot(1,5,i+1)
    plt.bar(df_clusters.columns, df_clusters.iloc[:, i], color='blue', alpha=0.2)
    plt.plot(df_clusters.columns, df_clusters.iloc[:,i], color='red')
    plt.title('Cluster ' + str(i))
    plt.xticks(rotation=45)
    plt.ylim(0,4)
```

```python
# PCA digunakan untuk melakukan visualisasi prediksi dalam grafik 2D
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_fit = pca.fit_transform(df)

df_pca = pd.DataFrame(data=pca_fit, columns=['PCA1', 'PCA2'])
df_pca['Clusters'] = predictions
df_pca.head()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data=df_pca, x='PCA1', y='PCA2', hue='Clusters', palette='Set2', alpha=0.8)
plt.title('Personality Clusters Using PCA')
plt.legend()
```
